```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Skiping first four rows contains metadata and loaded the dataset
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_85220.csv" , skiprows=4)
```

```python
df
```

```python
# droped two columns with null values
df.drop(columns=['2024','Unnamed: 69'], inplace=True)
```

```python
df.isnull().sum()
```

```python
# Drop rows with null values
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df.info()
```

```python
df = df[df['Indicator Name'] == 'Population, total']
```

```python
# Filtering only population rows
df_pop = df[df['Indicator Name'] == 'Population, total']

# Reshape from wide to long
df_long = df_pop.melt(
    id_vars='Country Name',
    value_vars=[str(year) for year in range(1960, 2024)],
    var_name='Year',
    value_name='Population'
)

# converting 'Year' to int
df_long['Year'] = df_long['Year'].astype(int)
```

```python
df_long
```

```python
india_pop = df_long[df_long['Country Name'] == 'India']
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(15, 8))
colors = sns.color_palette('viridis', n_colors=20)

# Barplot
ax = sns.barplot(
    data=india_pop,
    x='Year',
    y='Population',
    palette=colors
)

# Rotatating x-axis labels for readability
plt.xticks(rotation=90, ha='right', fontsize=11)
plt.yticks(fontsize=11)

# Adds gridlines for better visual guidance
ax.yaxis.grid(True)

# Add title and labels with custom font size
plt.title('Population Growth of India(1960-2023)', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=13)
plt.ylabel('Population', fontsize=13)


# showing bar labels with the gap of 10
for bar, label in zip(ax.patches, india_pop['Year']):
    if int(label) % 10 == 0:  # Showing only for 1960, 1970, ..., 2020
        x = bar.get_x() + bar.get_width() / 2
        y = bar.get_height()
        ax.text(
            x=x,
            y=y + y * 0.01,
            s=f"{int(y):,}",
            ha='center',
            va='bottom',
            fontsize=9,
            color='black'
        )

# Removing top and right spines for clean look
sns.despine()

plt.tight_layout()
plt.show()
```

```python
pop_2023 = df_long[df_long['Year'] == '2023']

plt.figure(figsize=(10, 6))
sns.histplot(data=pop_2023, x='Population', bins=20, kde=True, color='orange')

plt.title('Population Distribution Across Countries (2023)', fontsize=14, fontweight='bold')
plt.xlabel('Population')
plt.ylabel('Number of Countries')

plt.grid(True)
plt.tight_layout()
plt.show()
```
